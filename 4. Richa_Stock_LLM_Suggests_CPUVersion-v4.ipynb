```python
# ============================================================
# üß† Investment Research Multi-Agent System (CPU-Safe + Progress Bars)
# ============================================================

!pip install -q yfinance pandas numpy nltk transformers accelerate torch sentencepiece requests psutil matplotlib tqdm

import os, gc, json, torch, psutil, yfinance as yf, numpy as np, pandas as pd, requests, matplotlib.pyplot as plt
from transformers import pipeline
from datetime import datetime
import nltk
from tqdm import tqdm

nltk.download("punkt"); nltk.download("stopwords")

# ============================================================
# 1Ô∏è‚É£ Token Loader
# ============================================================
def safe_get_env(varname, prompt_text):
    val = os.getenv(varname)
    if val and len(val) > 8:
        return val
    try:
        from google.colab import userdata
        val = userdata.get(varname)
        if val:
            os.environ[varname] = val
            return val
    except Exception:
        pass
    val = input(f"üîë Enter your {prompt_text}: ").strip()
    os.environ[varname] = val
    return val

hf_token = safe_get_env("HUGGINGFACEHUB_API_TOKEN", "Hugging Face token")
news_key = safe_get_env("NEWS_API_KEY", "News API key")
print(f"‚úÖ HF token loaded: {hf_token[:10]}...")
print(f"‚úÖ News API key loaded: {news_key[:5]}...")

# ============================================================
# 2Ô∏è‚É£ Device + VRAM Detection (CPU-safe)
# ============================================================
if torch.cuda.is_available():
    device_id = 0
    device = torch.device("cuda")
    total_vram = torch.cuda.get_device_properties(0).total_memory / 1e9
    print(f"‚úÖ GPU detected: {torch.cuda.get_device_name(0)}, VRAM: {total_vram:.1f}GB")
else:
    device_id = -1
    device = torch.device("cpu")
    total_vram = psutil.virtual_memory().available / 1e9 * 0.7
    total_vram = min(total_vram, 8)  # force TinyLlama fallback
    print(f"‚ö†Ô∏è CPU detected, available RAM ~{total_vram:.1f}GB. TinyLlama will be used.")

print("Torch device count:", torch.cuda.device_count())

# ============================================================
# 3Ô∏è‚É£ Core Agents
# ============================================================
class BaseAgent:
    def log(self, msg): print(f"[{self.__class__.__name__}] {msg}")

class DataAgent(BaseAgent):
    def fetch_data(self, ticker, period="6mo"):
        self.log(f"Fetching data for {ticker}")
        data = yf.download(ticker, period=period, progress=False)
        if data.empty:
            self.log(f"‚ö†Ô∏è No data found for {ticker}, returning empty DataFrame")
        return data.reset_index()

class TechnicalAnalysisAgent(BaseAgent):
    def analyze(self, data):
        df = data.copy()
        if "Close" not in df.columns:
            raise ValueError("Data missing 'Close' column for technical analysis")
        df["SMA20"] = df["Close"].rolling(window=20).mean()
        df["SMA50"] = df["Close"].rolling(window=50).mean()
        delta = df["Close"].diff()
        gain = (delta.where(delta > 0, 0)).rolling(14).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
        rs = gain / loss
        df["RSI"] = 100 - (100 / (1 + rs))
        df = df.dropna()
        if df.empty: raise ValueError("Not enough data for analysis ‚Äî try longer period.")
        latest = df.iloc[-1]
        sma20, sma50, rsi = float(latest["SMA20"]), float(latest["SMA50"]), float(latest["RSI"])
        signal = "Bullish" if sma20 > sma50 else "Bearish"
        tech_score = np.clip((100 - abs(rsi - 50)) / 2, 0, 100)
        summary = {"SMA20":sma20,"SMA50":sma50,"RSI":rsi,"Signal":signal,"TechScore":tech_score}
        self.log(f"{signal} trend detected (RSI={rsi:.2f})")
        return summary

class RiskAssessmentAgent(BaseAgent):
    def assess(self, data):
        if "Close" not in data.columns:
            raise ValueError("Data missing 'Close' column for risk assessment")
        returns = data["Close"].pct_change().dropna()
        vol = float(returns.std()) if not returns.empty else 0.0
        drawdown = float((data["Close"] / data["Close"].cummax() - 1).min()) if not data.empty else 0.0
        risk_score = np.clip((vol + abs(drawdown)) * 100, 0, 100)
        summary = {"Volatility":vol,"Drawdown":drawdown,"RiskScore":risk_score}
        self.log(f"Risk Score: {risk_score:.2f}")
        return summary

class NewsSentimentAgent(BaseAgent):
    def __init__(self):
        self.api_key = os.getenv("NEWS_API_KEY")
        try:
            self.sentiment_pipe = pipeline(
                "text-classification",
                model="cardiffnlp/twitter-roberta-base-sentiment-latest",
                device=device_id,
                truncation=True,
                max_length=128,
            )
        except Exception as e:
            self.sentiment_pipe = None
            self.log(f"‚ö†Ô∏è Sentiment pipeline could not be loaded: {e}")

    def fetch_and_analyze(self, ticker):
        self.log(f"Fetching and scoring news for {ticker}")
        if not self.api_key or not self.sentiment_pipe:
            return "No News API or pipeline available.", 0.0
        url = f"https://newsapi.org/v2/everything?q={ticker}&language=en&sortBy=publishedAt&pageSize=5&apiKey={self.api_key}"
        r = requests.get(url)
        if r.status_code != 200: return "News API call failed.", 0.0
        arts = r.json().get("articles", [])
        if not arts: return "No recent articles.", 0.0
        texts = [a.get("title","") + " " + str(a.get("description", "")) for a in arts[:5]]
        preds = self.sentiment_pipe(texts)
        pos = sum(p["label"]=="positive" for p in preds)
        neg = sum(p["label"]=="negative" for p in preds)
        sentiment_score = ((pos - neg) / len(preds)) * 50 + 50  # Scale 0-100
        summary = "\n".join(texts)
        return summary, sentiment_score

class EvaluationAgent(BaseAgent):
    def evaluate(self, tech, risk, sentiment):
        if tech.get("Signal")=="Bullish" and risk.get("RiskScore",0)<40 and sentiment>55: return "BUY"
        elif tech.get("Signal")=="Bearish" and (risk.get("RiskScore",0)>60 or sentiment<45): return "SELL"
        else: return "HOLD"

# ============================================================
# 4Ô∏è‚É£ LLM Summarizer ‚Äî CPU-Safe + Progress
# ============================================================
def pick_model(vram):
    if vram > 28:
        return "mistralai/Mistral-7B-Instruct-v0.3"
    elif vram > 12:
        return "microsoft/Phi-2"
    else:
        return "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

def summarize_with_llm(draft, decision, alloc, news):
    model_name = pick_model(total_vram)
    print(f"üîß Using summarizer: {model_name} (CPU-safe)")
    result = draft
    try:
        # Nested progress bar for LLM generation
        from tqdm.auto import tqdm as tqdm_auto
        gen = pipeline(
            "text-generation",
            model=model_name,
            device=device_id,
            torch_dtype=torch.float32,
            max_new_tokens=512,
            temperature=0.5,
            repetition_penalty=1.1,
            trust_remote_code=True,
        )
        prompt = f"""
        You are a senior investment strategist. Rewrite this draft as a concise, professional
        200-word investment summary for clients. Integrate portfolio and news context.

        --- DRAFT ---
        {draft}

        --- DECISION ---
        {decision}

        --- PORTFOLIO ---
        {json.dumps(alloc)}

        --- NEWS ---
        {news}
        """
        # Show token generation progress
        for _ in tqdm_auto(range(1), desc="Generating summary"):
            result = gen(prompt, max_new_tokens=350, do_sample=True)[0]["generated_text"]
    except Exception as e:
        print(f"‚ö†Ô∏è LLM summarization failed: {e}\nReturning raw draft instead.")
    finally:
        del gen
        gc.collect()
        if torch.cuda.is_available(): torch.cuda.empty_cache()
    return result.strip()

# ============================================================
# 5Ô∏è‚É£ Orchestrator
# ============================================================
class InvestmentResearchAgent(BaseAgent):
    def __init__(self):
        self.data=DataAgent()
        self.tech=TechnicalAnalysisAgent()
        self.risk=RiskAssessmentAgent()
        self.news=NewsSentimentAgent()
        self.eval=EvaluationAgent()

    def run_ticker(self, ticker):
        data=self.data.fetch_data(ticker)
        if data.empty:
            return {"Ticker":ticker, "Decision":"DATA_ERROR"}
        try:
            tech=self.tech.analyze(data)
        except:
            tech={"SMA20":0,"SMA50":0,"RSI":0,"Signal":"UNKNOWN","TechScore":0}
        try:
            risk=self.risk.assess(data)
        except:
            risk={"Volatility":0,"Drawdown":0,"RiskScore":50}
        news_text, sentiment=self.news.fetch_and_analyze(ticker)
        decision=self.eval.evaluate(tech, risk, sentiment)
        draft=f"""
        Stock: {ticker}
        Signal: {tech.get('Signal','N/A')}
        RSI: {tech.get('RSI',0)}
        SMA20: {tech.get('SMA20',0)}
        SMA50: {tech.get('SMA50',0)}
        Volatility: {risk.get('Volatility',0):.4f}
        Drawdown: {risk.get('Drawdown',0):.2f}
        Risk Score: {risk.get('RiskScore',0):.2f}
        Sentiment Score: {sentiment:.2f}
        Recommendation: {decision}
        """
        optimized = summarize_with_llm(draft, decision, {}, news_text)
        return {"Ticker":ticker, **tech, **risk, "Sentiment":sentiment, "Decision":decision, "Summary":optimized}

# ============================================================
# 6Ô∏è‚É£ Run Pipeline Sequentially with Progress Bar
# ============================================================
ira = InvestmentResearchAgent()
tickers = ["AAPL","INTC","NVDA","MSFT","AMZN","TSLA","GOOGL"]
results = []

print("\nüìà Running Investment Research Pipeline:\n")
for t in tqdm(tickers, desc="Tickers processed"):
    try:
        res = ira.run_ticker(t)
        results.append(res)
    except Exception as e:
        print(f"\n‚ùå Error with {t}: {e}")
    finally:
        gc.collect()
        if torch.cuda.is_available(): torch.cuda.empty_cache()

df = pd.DataFrame(results)
df.to_csv("investment_summaries.csv", index=False)
print("\n‚úÖ All results saved to 'investment_summaries.csv'")

# ============================================================
# 7Ô∏è‚É£ Visualization - Radar Chart
# ============================================================
def plot_summary(df):
    if df.empty: return
    import matplotlib.pyplot as plt
    import numpy as np
    labels = ["TechScore", "RSI", "100-RiskScore", "Sentiment"]
    num_vars = len(labels)
    angles = np.linspace(0, 2*np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]
    plt.figure(figsize=(8,8))
    ax = plt.subplot(111, polar=True)
    for _, row in df.iterrows():
        values = [
            row.get("TechScore",0),
            row.get("RSI",0),
            100 - row.get("RiskScore",50),
            row.get("Sentiment",50)
        ]
        values += values[:1]
        ax.plot(angles, values, label=row.get("Ticker","N/A"))
        ax.fill(angles, values, alpha=0.1)
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    plt.xticks(angles[:-1], ["Tech", "RSI", "Safety", "Sentiment"])
    ax.set_rlabel_position(0)
    plt.title("üìä Stock Comparative Radar: Tech, RSI, Risk, Sentiment", size=14, weight='bold')
    plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
    plt.show()

plot_summary(df)

# ============================================================
# 8Ô∏è‚É£ Portfolio Weighting Logic
# ============================================================
def compute_weights(df):
    for col in ["TechScore","Sentiment","RiskScore"]:
        if col not in df.columns:
            df[col] = 0
    df["CompositeScore"] = (df["TechScore"] * 0.4) + (df["Sentiment"] * 0.4) + ((100 - df["RiskScore"]) * 0.2)
    df["Weight"] = np.where(df["Decision"]=="BUY", df["CompositeScore"], 0)
    total = df["Weight"].sum()
    df["Weight"] = (df["Weight"] / total * 100) if total>0 else 0
    df["Weight"] = df["Weight"].fillna(0).round(1)
    return df

df = compute_weights(df)

# ============================================================
# 9Ô∏è‚É£ Portfolio Summary
# ============================================================
def portfolio_summary(df):
    summary = []
    for _, row in df.iterrows():
        if row.get("Decision") == "BUY":
            summary.append(f"‚úÖ Buy {row.get('Ticker','N/A')} ‚Äî {row.get('Weight',0)}% weight")
        elif row.get("Decision") == "HOLD":
            summary.append(f"‚ö†Ô∏è Hold {row.get('Ticker','N/A')}")
        else:
            summary.append(f"‚ùå Avoid {row.get('Ticker','N/A')}")
    print("\nüìà Portfolio Recommendation Summary:\n" + "\n".join(summary))

portfolio_summary(df)

# ============================================================
# üîü Final Allocation Bar Chart
# ============================================================
plt.figure(figsize=(8,4))
buy_df = df[df["Decision"]=="BUY"]
plt.bar(buy_df.get("Ticker",[]), buy_df.get("Weight",[]))
plt.title("üìä Suggested Portfolio Allocation (%)")
plt.xlabel("Ticker"); plt.ylabel("Weight %")
plt.show()
```
